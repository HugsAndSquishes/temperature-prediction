```python
worst_country = hypertuning_results.loc[hypertuning_results["Best_MAE"].idxmax()]
print("Worst local model performance:")
print(worst_country)
print(worst_country["Best_Params"])
```

```python
scores = {}
for index, row in hypertuning_results.iterrows():
    #print(row["Scores"])
    country_scores = {}
    for label, score in ast.literal_eval(row["Scores"]).items():
        country_scores[label] = score
    scores[row["Country"]] = country_scores
```

```python
df = pd.DataFrame.from_dict(scores, orient='index')
df
```
